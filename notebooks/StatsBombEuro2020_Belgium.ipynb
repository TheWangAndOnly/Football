### Progressive Passes from Youri

```python
#scaling
Youri.startX = Youri.startX*1.2
Youri.endX = Youri.endX*1.2
Youri.startY = Youri.startY*.8
Youri.endY = Youri.endY*.8
```

I assume that a football pitch is 120 meters long and 40 meters large

```python
Youri['beginning'] = np.sqrt(np.square(120-Youri['startX']) + np.square(40-Youri['startY']))
Youri['end'] = np.sqrt(np.square(120-Youri['endX']) + np.square(40-Youri['endY']))
```

```python
Youri = Youri.drop(['location', 'pass_end_location'], axis =1)
```

__Did the ball progress at least 25% closer to the opposition goal?__

```python
#euclidian distance
Youri['progressive'] = [(Youri['end'].values[i]) / (Youri['beginning'].values[i]) <= .75 
                        for i in range(len(Youri['beginning']))]
```

```python
Youri['progressive'].value_counts()
```

__According to the euclidean distance I applied, Youri Tielemans only achieved 20 progressive passes__
